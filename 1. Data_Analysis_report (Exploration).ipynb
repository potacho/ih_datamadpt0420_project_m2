### Cleaning

We are going to remove entries with 0 in any of the x, y or z measures

Entries without width

```python
filter1=diamonds['y']==0
diamonds[diamonds['y']==0]
```

Entries without height

```python
filter2=diamonds['z']==0
diamonds[diamonds['z']==0]
```

Entries without length

```python
filter3=diamonds['x']==0
diamonds[diamonds['x']==0]
```

```python
diamonds=diamonds[~((filter1) | (filter2) | (filter3))]
```

New dataframe shape

```python
diamonds.shape
```
