```python
diamonds_cutypes=['Round', "Princess", "Cushion", "Oval", "Emerald", "Heart", "Pear", "Marquise", "Asscher","Radiant"]
d=dict()
for i in diamonds_cutypes:
    url=f'https://www.brilliance.com/education/diamond-size-chart/{i}'
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find_all('td')  
    d[i]=[table[x:x+2] for x in range(0,len(table),2)]
```

```python
def clean(x):
    return re.sub('<.*?>','',str(x))
```

```python
def clean1 (x):
    return re.sub('ct.','',str(x))
```

```python
def clean2 (x):
    return re.sub('mm.','',str(x))
```

```python
x=d["Princess"]
length=[]
carat=[]
for i in x:
    length.append(i[0])
    carat.append(i[1])
princess=pd.DataFrame()
princess['Princess length']=length
princess['carat']=carat
princess['Princess length']=princess['Princess length'].apply(clean)
princess['carat']=princess['carat'].apply(clean)
princess['Princess length']=princess['Princess length'].apply(clean2)
princess['carat']=princess['carat'].apply(clean1)
```

```python
princess['Princess length']=princess['Princess length'].astype('float64', copy=False)
princess['carat']=princess['carat'].astype('float64', copy=False)
diamonds= pd.merge(diamonds,princess, on='carat', how='left')
diamonds['Princess length']=diamonds['Princess length'].fillna(0)
```

```python
diamonds['diamonds_type']=np.where(diamonds['Princess length']==diamonds['x'], 'princess', 'other')
```
