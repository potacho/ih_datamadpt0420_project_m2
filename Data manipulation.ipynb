```python
diamonds_cutypes=['Round', "Princess", "Cushion", "Oval", "Emerald", "Heart", "Pear", "Marquise", "Asscher","Radiant"]
d=dict()
for i in diamonds_cutypes:
    url=f'https://www.brilliance.com/education/diamond-size-chart/{i}'
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find_all('td')  
    d[i]=[table[x:x+2] for x in range(0,len(table),2)]
```

```python
def clean(x):
    return re.sub('<.*?>','',str(x))
```

```python
def clean1 (x):
    return re.sub('ct.','',str(x))
```

```python
def clean2 (x):
    return re.sub('mm.','',str(x))
```

```python
round=pd.DataFrame()
x=d['Round']
width=[]
carat=[]
for i in x:
    width.append(i[0])
    carat.append(i[1])
round['Round width']=width
round['carat']=carat
round['Round width']=round['Round width'].apply(clean)
round['carat']=round['carat'].apply(clean)
round['Round width']=round['Round width'].apply(clean2)
round['carat']=round['carat'].apply(clean1)
```

```python
round['Round width']=round['Round width'].astype('float64', copy=False)
round['carat']=round['carat'].astype('float64', copy=False)
```

```python
diamonds= pd.merge(diamonds,round, on='carat').fillna(0)
```
